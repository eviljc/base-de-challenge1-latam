```python
import pandas as pd

# Paso 1: Definir los URLs y nombres de las tiendas en un diccionario.
tiendas_urls = {
    'Tienda_1': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    'Tienda_2': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    'Tienda_3': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    'Tienda_4': "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

# Paso 2: Usar una lista de comprensión para cargar y concatenar los datos de forma concisa.
df_list = []
for tienda, url in tiendas_urls.items():
    df_temp = pd.read_csv(url)
    df_temp['Tienda'] = tienda
    df_list.append(df_temp)
df = pd.concat(df_list, ignore_index=True)

# Paso 3: Agrupar por tienda y producto para contar las ventas.
producto_ventas = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad Vendida')

# Paso 4: Encontrar los productos más y menos vendidos para cada tienda de forma eficiente.
idx_most_sold = producto_ventas.groupby('Tienda')['Cantidad Vendida'].idxmax()
idx_least_sold = producto_ventas.groupby('Tienda')['Cantidad Vendida'].idxmin()

most_sold_df = producto_ventas.loc[idx_most_sold]
least_sold_df = producto_ventas.loc[idx_least_sold]

# Paso 5: Mostrar los resultados de forma clara y concisa.
print("--- Productos Más y Menos Vendidos por Tienda ---")
for tienda in tiendas_urls.keys():
    most_sold_product = most_sold_df[most_sold_df['Tienda'] == tienda].iloc[0]
    least_sold_product = least_sold_df[least_sold_df['Tienda'] == tienda].iloc[0]

    print(f"\n--- {tienda} ---")
    print(f"Producto más vendido: {most_sold_product['Producto']} (Cantidad: {most_sold_product['Cantidad Vendida']})")
    print(f"Producto menos vendido: {least_sold_product['Producto']} (Cantidad: {least_sold_product['Cantidad Vendida']})")

```

```python
import pandas as pd

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda['Tienda'] = 'Tienda_1'
tienda2['Tienda'] = 'Tienda_2'
tienda3['Tienda'] = 'Tienda_3'
tienda4['Tienda'] = 'Tienda_4'


df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)


productos_tienda = df.groupby(['Tienda', 'Producto'])['Precio'].agg(['sum', 'count'])


productos_tienda.rename(columns={'sum': 'Ganancia', 'count': 'Cantidad'}, inplace=True)


productos_tienda
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

productos_tienda_plot = productos_tienda.reset_index()

ax = productos_tienda_plot.groupby('Tienda').sum().plot(kind='bar', stacked=True, y=['Ganancia'])
cmap = plt.colormaps['viridis']


norm = plt.Normalize(0.2, 0.8)

# Apply the colormap to the bars
for i, bar in enumerate(ax.patches):
    bar.set_facecolor(cmap(norm(i)))

ax.bar_label(ax.containers[0], fmt='{:,.0f}', label_type='edge')
plt.xlabel('Tienda')
plt.ylabel('Valores')
plt.title('Ganancia por Tienda')
plt.xticks(rotation=45)
plt.legend(title='Métrica')
plt.tight_layout()
plt.show()
```
